```python
import numpy as np
import math
import pandas as pd
from statistics import mean, stdev
from tqdm.notebook import trange, tqdm
import warnings
warnings.simplefilter('ignore')

from social_force_model import SocialForceModel
from destination_choice_model import DestinationChoiceModel

import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
import matplotlib.animation as animation
%matplotlib notebook
plt.rcParams['animation.ffmpeg_path'] = '/usr/bin/ffmpeg'
```

```python
## Agent
class Agent():
    def __init__(self, no, t0, r, v0, loc, vel, dest, type_):
        self.no = no
        self.t0 = t0 
        self.r = r
        self.v0 = v0 
        self.loc = loc
        self.vel = vel
        self.dest = dest
        self.type = type_
        self.done = False
```

```python
def run(ped_info, num_step, sfm, dcm, real_trajectories=None):
    
    # initialize agents
    num_agent = len(ped_info)
    agents = {}
    for i in range(num_agent):
        agents[i] = Agent(no = ped_info.ID[i],
                          t0 = ped_info.frame_s[i],
                          r = 0.2,
                          v0 = ped_info.speed_mean[i],
                          loc = [np.array([np.nan, np.nan])],
                          vel = [np.array([np.nan, np.nan])],
                          dest = [np.array([np.nan, np.nan])],
                          type_ = ped_info.type[i])

    t = 0
    while t < num_step: # until all pedestrians move to right
        for i in range(num_agent):
            # generate agents
            if agents[i].t0 == t:
                agents[i].loc[t] = np.array([ped_info.start_x[i], ped_info.start_y[i]])
                agents[i].vel[t] = np.array([ped_info.speed_mean.mean(), 0])
                agents[i].dest[t] = dests[0]

            # real
            if (agents[i].type == "real"):
                agents[i].loc.append(np.array([real_trajectories[i].x[t], real_trajectories[i].y[t]]))
                agents[i].vel.append(np.array([real_trajectories[i].vel_x[t], real_trajectories[i].vel_y[t]]))
                if real_trajectories[i].choice[t] - 1 in [0, 1]:
                    agents[i].dest.append(dests[int(real_trajectories[i].choice[t]-1)])
                else:
                    agents[i].dest.append(np.array([np.nan, np.nan]))

            # model
            elif (agents[i].t0 <= t) and (agents[i].done == False): 
              
                loc = agents[i].loc[t]
                vel = agents[i].vel[t]
                if t % 15 == 0:
                    dest = dcm.choice(t, i, agents)
                else:
                    dest = agents[i].dest[t]
                    
                loc_next, vel_next = sfm.step(t, i, agents, dest)
                
                agents[i].loc.append(loc_next)
                agents[i].vel.append(vel_next)
                agents[i].dest.append(dest)
                
                if np.linalg.norm(loc_next - dest) <= 0.2:
                    agents[i].done = True
            else:
                agents[i].loc.append(np.array([np.nan, np.nan]))
                agents[i].vel.append(np.array([np.nan, np.nan]))
                agents[i].dest.append(np.array([np.nan, np.nan]))
        t += 1
    
    return agents
```

```python
def visualization(agents, dests, walls, num_step, output_path):
    
    num_agent = len(agents)

    ppi=72 # points per inche 

    xmin, xmax = 0, 2.5 
    ymin, ymax = -1.75, 1.5

    fig = plt.figure(figsize=(32/4, 17/4), dpi=100.0)
    ax = plt.axes()
    ax.set_aspect("equal")
    ax.axis("off")


    ax_length=ax.bbox.get_points()[1][0]-ax.bbox.get_points()[0][0]
    ax_point = ax_length*ppi/fig.dpi
    xsize=xmax-xmin
    fact=ax_point/xsize

    plt.tick_params(labelsize=12)

    #
    title = ax.text(0.34, 1.05, '', transform=ax.transAxes, va='center', fontsize=16)
    point = []
    ana = [] 

    for j in range(num_agent):
        # points, = plt.scatter([], [], s=(r[j]*2*fact)**2, color="red", linewidths=0) # color=colors[j%10])
        if agents[j].type == "real":
            points, = plt.plot([], [], 'o', c="#1f77b4", markersize=agents[j].r*2*16)
        else:
            points, = plt.plot([], [], 'o', c="black", markersize=agents[j].r*2*16)
    #     points, = plt.plot([], [], 'o', c="black", markersize=R*2*16)
        point.append(points)

    point = tuple(point)

    def init():
        for j in range(num_agent):
            point[j].set_data([],[])
            title.set_text('')
        return point, title

    def animate(i):
        for h,a in enumerate(ana):
            a.remove()
        ana[:] = []

        for j in range(num_agent):
            if 1 * (sum(agents[j].dest[i] == dests[0]) == 2):
                ana_j = plt.annotate("→",(agents[j].loc[i][0]+0.05, agents[j].loc[i][1]), color='black',size=8)
            else:
                ana_j = plt.annotate("↑",(agents[j].loc[i][0]+0.05, agents[j].loc[i][1]), color='black',size=8)
            ana.append(ana_j)
            point[j].set_data(agents[j].loc[i][0], agents[j].loc[i][1])
        title.set_text('t=' + str(int(i/30)))
        return point, title, ana

    anim = animation.FuncAnimation(fig, animate, init_func=init, frames=num_step, interval=30, blit=True)
    ax.set_xlim(xmin,xmax)
    ax.set_ylim(ymin,ymax)

    for i in range(len(walls))    :
        plt.plot(walls[i][:, 0], walls[i][:, 1], c="black")
#     plt.plot([0, 2.5], [-1.75, -1.75], c="black", linewidth=3)
#     plt.plot([0, 2.5], [1.5, 1.5], c="black", linewidth=3)
#     plt.plot([0, 0], [-1.75, 1.5], c="black", linewidth=3)
#     plt.plot([2.5, 2.5], [-1.75, 1.5], c="black", linewidth=3)

    # plt.scatter(dests[0][0], dests[0][1])
    # plt.scatter(dests[1][0], dests[1][1])

    anim.save(output_path, writer="ffmpeg")
#     mywriter = animation.FFMpegWriter()
```

# main

```python
# environment
walls = [
    np.array([(0.3, y) for y in np.linspace(0.5, 3.5, 30)]),
    np.array([(0.3, y) for y in np.linspace(-1.6, -2.4, 8)]),
    np.array([(1.8, y) for y in np.linspace(1.1, 3.5, 24)]),
    np.array([(2.1, y) for y in np.linspace(0.4, 1.1, 7)]),
    np.array([(2.1, y) for y in np.linspace(-1.2, -2.4, 12)]),
    np.array([(x, 0.5) for x in np.linspace(-0.9, 0.3, 12)]),
    np.array([(x, -1.6) for x in np.linspace(-0.9, 0.3, 12)]),
    np.array([(x, 1.1) for x in np.linspace(1.8, 2.1, 3)]),
    np.array([(x, 0.4) for x in np.linspace(2.1, 5.0, 29)]),
    np.array([(x, -1.2) for x in np.linspace(2.1, 5.0, 29)]),
]
walls_points = []
for wall in walls:
    walls_points.extend(list(wall))  
walls_points = list(map(list, set(map(tuple, walls_points))))

threshold = [
    np.array([[0, -1.75], [2.5, -1.75]]),
    np.array([[0, 1.5], [2.5, 1.5]]),
    np.array([[0, -1.75], [0, 1.5]]),
    np.array([[2.5, -1.75], [2.5, 1.5]])
]

walls.extend(threshold)

dests = [np.array([5, -0.4]), np.array([1.05, 5])]
```

```python
# measurement data
ped_info = pd.read_csv("../data/departure_time_and_initial_position/measurement_data.csv")
ped_info.speed_mean = ped_info.speed_mean.fillna(ped_info.speed_mean.mean())

ped_info["type"][0] = "real"
ped_info["type"][2] = "real"
ped_info["type"][3] = "real"

# actual trajectiries used for simulation
real_trajectory_id2 = pd.read_csv("../data/trajectory/each_pedestrian/_ID2.csv")  
real_trajectory_id4 = pd.read_csv("../data/trajectory/each_pedestrian/_ID4.csv")  
real_trajectory_id5 = pd.read_csv("../data/trajectory/each_pedestrian/_ID5.csv")  
real_trajectory_list = [real_trajectory_id2, real_trajectory_id4, real_trajectory_id5]

for real_trajectory in real_trajectory_list:
    vel_x = [np.nan]
    vel_y = [np.nan]
    for i in range(1, len(real_trajectory)):
        vel_x.append((real_trajectory.x[i] - real_trajectory.x[i-1]) / params_sfm["dt"])
        vel_y.append((real_trajectory.y[i] - real_trajectory.y[i-1]) / params_sfm["dt"])
    real_trajectory["vel_x"] = vel_x
    real_trajectory["vel_y"] = vel_y
    
real_trajectories = {}
real_trajectories[0] = real_trajectory_id2 
real_trajectories[2] = real_trajectory_id4
real_trajectories[3] = real_trajectory_id5
```

```python
# parameters of SFM
params_sfm = {
    "dt": 1/30,
    "A1": 2.1,
    "B": 0.3,
    "A2": 1,
    "tau": 0.5,
    "phi": 100,
    "c": 0.5
}
```

```python
num_step = 1600
```

# Shortest path

```python
sfm = SocialForceModel(params_sfm, walls_points)

# parameters of DCM
params_dcm = {
    "ASC_2": 0,
    "distance": -1.33,
    "previous_choice": 0,
    "others_choice_front": 0,
    "others_choice_behind": 0
}
dcm = DestinationChoiceModel(params_dcm, dests)
```

```python
# run simulation 50 times
num_r1_list = []
num_r2_list = []
for trial in trange(50):
    
    agents = run(ped_info, num_step, sfm, dcm, real_trajectories)
    
    if trial == 0:
        visualization(agents, dests, walls, num_step, "./simulated_movement_shortest_path.mp4")

    num_r1 = 0
    num_r2 = 0
    for i in range(len(agents)):
        data = pd.DataFrame(agents[i].loc)[0]
        if data[data.notna()][-1:].values > 2.0:
            num_r1 += 1
        else:
            num_r2 += 1
            
    num_r1_list.append(num_r1)
    num_r2_list.append(num_r2)
    
num_r1_list_sp = num_r1_list
num_r2_list_sp = num_r2_list
```

# DCM

```python
# parameters of DCM
params_dcm = {
    "ASC_2": 2.33,
    "distance": -1.33,
    "previous_choice": 1.13,
    "others_choice_front": 0.202,
    "others_choice_behind": -0.105
}
dcm = DestinationChoiceModel(params_dcm, dests)
```

```python
# run simulation 50 times
num_r1_list = []
num_r2_list = []
for trial in trange(50):
    
    agents = run(ped_info, num_step, sfm, dcm, real_trajectories)
    
    if trial == 0:
        visualization(agents, dests, walls, num_step, "./simulated_movement_dcm.mp4")

    num_r1 = 0
    num_r2 = 0
    for i in range(len(agents)):
        data = pd.DataFrame(agents[i].loc)[0]
        if data[data.notna()][-1:].values > 2.0:
            num_r1 += 1
        else:
            num_r2 += 1
            
    num_r1_list.append(num_r1)
    num_r2_list.append(num_r2)
    
num_r1_list_dcm = num_r1_list
num_r2_list_dcm = num_r2_list
```

# Aggregation of results

```python
err_sp = [stdev(num_r1_list_sp), stdev(num_r2_list_sp)]
err_dcm = [stdev(num_r1_list_dcm), stdev(num_r2_list_dcm)]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.bar([0,2], [6, 47], width=0.5, color="gray", label="Real")

plt.bar([0.5,2.5], [mean(num_r1_list_dcm), mean(num_r2_list_dcm)], width=0.5, yerr=err_dcm, capsize=5, label="SP")
plt.bar([1,3], [mean(num_r1_list_sp), mean(num_r2_list_sp)], width=0.5, yerr=err_sp, capsize=5, label="DCM")
plt.ylabel("Number of pedestrians who chose the route")
plt.xticks([0.5, 2.5], ["Route 1", "Route 2"])
plt.legend()
plt.savefig("../outputs/num_ped_chose_route.png", bbox_inches="tight")
```
